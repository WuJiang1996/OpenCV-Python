```python
import cv2 as cv
import numpy as np

img = cv.imread('messi.png',0)
cv.imshow('image',img)
k = cv.waitKey(0)
if k == 27:      #等待esc退出
    cv.destroyAllWindows()
elif k == ord('s'):    #等待关键字，保存和退出
    cv.imwrite('messigray.png',img)
    cv.destroyAllWindows()
```
